```python
#cleaning the data
df['ssc_p']=df['ssc_p'].fillna(df['ssc_p'].mean())
df['ssc_b']=df['ssc_b'].fillna(df['ssc_b'].mode()[0])
df['hsc_s']=df['hsc_s'].fillna(df['hsc_s'].mode()[0])
df['degree_p']=df['degree_p'].fillna(df['degree_p'].mean())
df['degree_t']=df['degree_t'].fillna(df['degree_t'].mode()[0])
df['workex']=df['workex'].fillna(df['workex'].mode()[0])
df['etest_p']=df['etest_p'].fillna(df['etest_p'].mean())
df['specialisation']=df['specialisation'].fillna(df['specialisation'].mode()[0])
# Drop rows where salary is missing
df = df.dropna(subset=["salary"])
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

```python
df.isnull().sum()
```

```python
x=df.drop(columns=['salary'])
x
```

```python
y=df['salary']
y
```

```python
string_cols = df.select_dtypes(include='object').columns
string_cols
df = pd.get_dummies(df, columns=string_cols, drop_first=True)
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
pd.DataFrame({'y_test':y_test,'y_pred':y_pred,'MAE':y_test-y_pred})
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
# --- METRICS ---
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
MAE = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", MAE)
print("R² Score:", r2)
```
