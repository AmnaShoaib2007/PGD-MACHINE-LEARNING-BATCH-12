```python
#df=pd.get_dummies(df) is called One Hot Encoding
df['has_covid'] = df['has_covid_Yes'].apply(lambda x: 'Yes' if x == 1 else 'No')
df = df.drop(columns=['has_covid_No', 'has_covid_Yes'])

df
```

```python
df.columns
```

```python
x=df.drop(columns=['has_covid'])
x
```

```python
y=df['has_covid']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=45)
x_train
```

```python
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
y_pred
```

```python
pd.DataFrame({'y_test':y_test,'y_predict':y_pred})
```

```python
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix,recall_score,classification_report
print ("Accuracy Score:",accuracy_score(y_test,y_pred))
print("accuracy_score:",accuracy_score(y_test,y_pred))
print("confusion_matrix:",confusion_matrix(y_test,y_pred))
print("precision_score:",precision_score(y_test,y_pred,pos_label="Yes"))
print("recall_score:",recall_score(y_test,y_pred,pos_label="Yes"))
print("classification_report: \n",classification_report(y_test,y_pred,))
```
