```python
cols_to_drop = ['id','City','Profession','Degree']
df = df.drop(columns=[col for col in cols_to_drop if col in df.columns])
df
```

```python
numeric_df=df.select_dtypes(include=[np.number])
numeric_df.isnull().sum()
```

```python
for col in numeric_df:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].mean(), inplace=True)
numeric_df
    
```

```python
for col in numeric_df:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].median(), inplace=True)
numeric_df
    
```

```python
numeric_df.isnull().sum()
```

```python
cat_df=df.select_dtypes(include='object')
for col in cat_df:
    if df[col].dtype in ['object','category']:
        df[col]=df[col].fillna(df[col].mode()[0])
cat_df
```

```python
cat_df.isnull().sum()
```

```python
df.isnull().sum()[df.isnull().sum()>0]
```

```python
df=pd.get_dummies(df,drop_first=True)
df
```

```python
df.head()
```

```python
x=df.drop(columns=['Depression'])
x
```

```python
y=df['Depression']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
pd.DataFrame({'y_test':y_test, 'y_predict':y_pred})
```

```python
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix,recall_score,f1_score,classification_report
print("Accuracy Score:",accuracy_score(y_test,y_pred))
print("Precision Score:",precision_score(y_test,y_pred))
print("Recall Score:",recall_score(y_test,y_pred))
print("F1 Score:",f1_score(y_test,y_pred))
```

```python
print("Confusion Matrix:",confusion_matrix(y_test,y_pred))
print("Classification Report:\n",classification_report(y_test,y_pred))
```
